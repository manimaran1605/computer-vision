```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'C:\Users\muthu\Downloads\test_img.png')
```

```python
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
edges = cv2.Canny(image_rgb, 400, 1)
```

```python
edges_inverted = 255 - edges  # only if edges are white-on-black

fig, axs = plt.subplots(1, 2, figsize=(7, 4))

# Show original RGB image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')

# Show edge image with white background and black edges
axs[1].imshow(edges_inverted, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Image Edges (Black on White)')

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```
