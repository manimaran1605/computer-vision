```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r'C:\Users\muthu\Downloads\classification_images\dogs\dog_5.png')
```

```python
h, w = image.shape[:2]
print("Height = {}, Width = {}".format(h,w))
```

```python
(B,G,R) = image[100,100]
print("R = {}, G = {}, B = {}".format(R, G, B))
```

```python
B = image[100, 100, 0]
print("B = {}".format(B))
G = image[100, 100, 1]
print("G = {}".format(G))
```

```python
roi = image[100:500, 200: 700]
image_rgb = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB) #opencv uses BGR, convert this to RGB
```

```python
plt.imshow(image_rgb)
plt.axis('off')  # Hide axis
plt.show()
```
